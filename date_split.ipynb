```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('./订单时间数据.xlsx')
```

```python
#转成datatime格式
df['Time'] = pd.to_datetime(df['Time'])
df.sort_values('Time',inplace=True)
df
```

```python
df_1 = df.set_index('Time')
df_1
```

```python
# 取某一天的数据
df_1['2019-09-09'].reset_index('Time')
```

```python
#按时间段切分 输入起止日期t1,t2
t1 = '2019-09-09 07'    
t2 = '2019-09-09 09'
df_1.truncate(before = t1, after = t2)
```
